```python
url = "https://www.aphis.usda.gov/sites/default/files/frequent-foia-requests.csv"
url
```

```python
df = pd.read_csv(url)
df
```

```python
## extract the url from the href in title
## I used AI to help me with this
df["file_url"] = df["Title"].apply(
    lambda x: BeautifulSoup(x, "html.parser").find("a")["href"] if x else None
)
```

## Text Analysis

```python
## download all documents
import wget
```

```python
## download all documents, used AI to assist
base_url = "https://www.aphis.usda.gov"

df["file_url_full"] = df["file_url"].apply(
    lambda x: base_url + x if x else None
)


for link in df["file_url_full"]:
    if link:
        wget.download(link)
        time.sleep(uniform(5, 10))
```

```python
## This froze after downloading 107 files, out of 139, so I used Chatgbt to help me figure out how to download the remaining 32 files
import os
```

```python
## This froze after downloading 107 files, out of 139, so I used Chatgbt to help me figure out how to download the remaining 32 files
for url in df["file_url_full"]:
    if not url:
        continue

    filename = url.split("/")[-1]

    # Skip if file already exists in current directory
    if os.path.isfile(filename):
        print(f"Skipping already downloaded: {filename}")
        continue

    print(f"Downloading: {filename}")
    try:
        wget.download(url)
        print("\Done")
    except Exception as e:
        print(f"\ERROR downloading {filename}: {e}")

    time.sleep(uniform(5, 10))
```
